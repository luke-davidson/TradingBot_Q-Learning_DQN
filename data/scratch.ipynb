```python
from os import listdir
from os.path import isfile, join

import numpy as np

from utils.experiment import ExperimentResult
from utils.plotting import plot_curves
```

```python
path = '../data/'
filenames = sorted([f for f in listdir(path) if isfile(join(path, f)) and f[-7:] == '.pickle'])
```

```python
filenames
```

```python
files_stats = {}
for filename in filenames:
    try:
        results = ExperimentResult.from_file(join(path, filename))
        files_stats[filename] = {
            'profit_length': len(results.profits),
            'returns_length': len(results.returns) if hasattr(results, 'returns') else 0,
        }
    except Exception as e:
        print(e)
        print(f'Could not load pickle file {filename}')
```

```python
print(f'name \t profits length \t returns length')
for name, stats in files_stats.items():
    print(f'{name} \t {stats["profit_length"]} \t {stats["returns_length"]}')
```

```python



selected = 'dqn_2022-11-21--03-29-02.pickle'
results = ExperimentResult.from_file(join(path, selected))
print('length:', len(results.profits))
print(results.profits)
```

```python
def moving_average(a, n=3):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

plot_curves([np.array([results.profits]), np.array([(moving_average(results.profits, n=5))])],
                ['Profits', '5-episode moving average'], ['r', 'g'], xlabel='Episode', ylabel='Profit ratio',
                title='DQN')
```
