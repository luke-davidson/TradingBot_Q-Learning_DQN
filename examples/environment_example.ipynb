```python
import matplotlib.pyplot as plt
from easydict import EasyDict

import environment
```

```python
env = environment.StocksEnv(EasyDict({
    "env_id": 'stocks-v0', "eps_length": 30,
    "window_size": 10, "train_range": None, "test_range": None,
    "stocks_data_filename": 'STOCKS_GOOGL'
}))
env.seed(0)
```

```python
env.reset()
```

```python
# from environment import Action

state = env.reset()
print(f'State: {state}')
for i in range(5):
    print(f'Time step {i} ===================================================')
    action = env.random_action()
    print(f'{action} Action: {environment.Action(action).name}')
    step = env.step(action)

    print(f'Next state: {step.obs}')
    print(f'Reward: {step.reward}')
```
